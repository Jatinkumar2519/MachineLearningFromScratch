```python
import numpy as np
import seaborn as sns
from sklearn.datasets import make_blobs
```

```python
features,target = make_blobs(n_samples=500,n_features=2,cluster_std=[1,1],centers=[[4,4],[-2,-2]])
```

```python
sns.scatterplot(x=features[:,0],y=features[:,1],hue=target)
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,20):
    km = KMeans(n_clusters=i)
    km.fit(features,target)
    wcss.append(km.inertia_)
```

```python
sns.pointplot(x=range(1,20),y = wcss)
```

```python
from sklearn.metrics import accuracy_score
model = KMeans(n_clusters=2)
model.fit(features,target)
accuracy_score(model.predict(features),target)
```
