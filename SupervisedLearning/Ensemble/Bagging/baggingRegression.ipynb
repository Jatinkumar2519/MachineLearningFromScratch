```python
import numpy as np
import pandas as pd
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
df = pd.read_csv("D://Programming//Machine_Learning//MachineLearningAlgo//SupervisedLearning//DataSets//advertising.csv")
```

```python
target = df.iloc[:,1]
features = df.iloc[:,:-1]
le = LabelEncoder()
target = le.fit_transform(target.values)
sc = StandardScaler()
features = sc.fit_transform(features)
```

```python
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
estimators = [("lr",LinearRegression()),
             ("dt",DecisionTreeRegressor(max_depth=50)),
             ("knn",KNeighborsRegressor(n_neighbors=5)),
             ("svm",SVR(kernel="sigmoid"))]
```

```python
for model in estimators:
    print(f"{model[0]} accuracy:",np.mean(cross_val_score(model[1],features,target,cv = 10)))
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
for model in estimators:
    bg = BaggingRegressor(estimator=model[1],bootstrap=True,n_estimators=50,random_state=42)
    print(f"with base model {model[0]} accuracy:",np.mean(cross_val_score(bg,features,target,cv = 10)))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=50,max_depth=50)
print("with base model rf accuracy:",np.mean(cross_val_score(rf,features,target,cv = 10)))
```
