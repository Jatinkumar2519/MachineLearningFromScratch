```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
class GradientBoost:
    def __init__(self,learning_rate = 0.1,n_estimators = 10,max_depth = 1):
        self.learning_rate = learning_rate
        self.n_estimators = n_estimators
        self.max_depth = max_depth
    
    def fit(self, x, y):
        y = y.copy()
        pred = y.mean()
        self.init = pred
        y -= pred
        self.dts = []
        for _ in range(self.n_estimators):
            model = DecisionTreeRegressor(max_depth=self.max_depth)
            model.fit(x, y)
            y -= self.learning_rate * model.predict(x)
            self.dts.append(model)

    
    def predict(self, x):
        pred = np.full(x.shape[0], self.init)
        for model in self.dts:
            pred += self.learning_rate * model.predict(x)
        return pred

        
```

```python
df = pd.read_csv("D://Programming//Machine_Learning//MachineLearningAlgo//SupervisedLearning//DataSets//advertising.csv")
features = df.iloc[:,:3]
target = df["Sales"]
```

```python
x_train, x_test,y_train,y_test = train_test_split(features,target,random_state=42,test_size=0.2)
```

```python
model = GradientBoost(n_estimators=150,max_depth=3)
model.fit(x_train.values,y_train.values)
r2_score(y_test,model.predict(x_test.values))
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor(n_estimators=150,max_depth=3,learning_rate=0.1)
model.fit(x_train.values,y_train.values)
r2_score(y_test,model.predict(x_test.values))
```
