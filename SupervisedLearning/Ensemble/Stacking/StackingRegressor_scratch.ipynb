```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(
"D://Programming//Machine_Learning//MachineLearningAlgo//SupervisedLearning//DataSets//" \
"advertising.csv")
features = df.iloc[:,:-1].values
target = df.iloc[:,-1].values.reshape(-1,1)
```

```python
normalize = StandardScaler()
features = normalize.fit_transform(features)
target = normalize.fit_transform(target)
```

```python
class Stacking:
    def __init__(self,base_estimators,meta_estimator):
        self.base_estimators = base_estimators
        self.meta_estimator = meta_estimator
    
    def fit(self,x,y):
        df = pd.DataFrame()
        for name,model in self.base_estimators:
            model.fit(x,y)
            df[name] = model.predict(x).flatten()
        
        self.meta_estimator.fit(df.values,y)
    
    def predict(self,x):
        df = pd.DataFrame()
        for name,model in self.base_estimators:
            df[name] = model.predict(x).flatten()
        
        return self.meta_estimator.predict(df.values)
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.2,random_state=42)
```

```python
estimators = [
    ("knn",KNeighborsRegressor(n_neighbors=5)),
    ("dt",DecisionTreeRegressor(max_depth=10))
]
meta = LinearRegression()
```

```python
model = Stacking(base_estimators=estimators,meta_estimator=meta)
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,model.predict(x_test))
```
